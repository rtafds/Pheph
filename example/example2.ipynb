### This library is to find the best explanatory variables.
- First, learning the relationship between explanatory variables and objective variables.
- Second, find an explanatory variable that maximizes the evaluation value created from the objective variable in GA

```python
import pandas as pd
import numpy as np
from pheph import Pheph
```

```python
pheph = Pheph()
```

```python
df = pd.read_csv("../datasets/adult_raw.csv")
```

```python
#df_fillna = pheph.complena(df)  # now, error.
df_dropna = df.dropna()
```

```python
df_dropna
```

```python
# dummies and categories. if don't use this function, after can't restore daga from generating GA
df_formatting, dummies_list, categories_reborn, category_colnames = pheph.formatting(df_dropna, dummies=["workclass","education","marital-status","occupation", "relationship", "race","native-country"],category_le_list=["sex","class"])
```

```python
df_formatting.iloc[:,:]
```

```python
# data reverse
df_inverse = pheph.inverse_formatting(df_formatting)
```

```python
df_inverse
```

```python
n_col = df_formatting.shape[1]
```

```python
exp = [x for x in range(n_col-1)]
obj = [n_col-1]
algorithms = "GradientBoostingRegressor"

#algorithms = 'GaussianProcessRegressor'
```

```python
# learning 
pheph.fit(df_formatting, exp, obj, algorithms=algorithms)
```

```python
# search best explanation
# this case is single optimization
best_ind, best_ind_value, pop, pop_save = pheph.inverse_fit(df_formatting)
```

```python
best_ind
```

```python
best_ind_value
```

```python
pop_save
```

```python
# from pop_save from GA and raw_data, sort from those with the highest rating
fit_data = pheph.ga_restore(pop_save, df_formatting)  
```

```python
# ga_restore processing
# from pop_data, add values
pop_fit_data = pheph.pop_data_fitness_value(pop_save, weights='self', ga_function='self')
```
